````python
#Function edited to make sure we don t find rows that contain substrings of given words.
def filter_questions(df, words):
    # Convert the words list to lowercase
    words = [word.lower() for word in words]
    # Create a mask with True/False values for each row
    mask = df['Question'].apply(lambda x: all(word in x.lower().split() for word in words))
    # Return the filtered DataFrame
    return df[mask]
````

5. Write a function that returns the count of unique answers to all of the questions in a dataset. For example, after filtering the entire dataset to only questions containing the word `"King"`, we could then find all of the unique answers to those questions. The answer "Henry VIII" appeared 55 times and was the most common answer.

````python
def count_unique_answers(df, keyword):
    # Filter the dataframe to only include rows where the question contains the keyword
    filtered_df = df[df['Question'].str.contains(keyword, case=False, na=False)]
    
    # Get the count of unique answers
    unique_answers = filtered_df['Answer'].nunique()
    
    # Get the most common answer and its count
    most_common_answer = filtered_df['Answer'].value_counts().idxmax()
    most_common_count = filtered_df['Answer'].value_counts().max()
    
    return unique_answers, most_common_answer, most_common_count

# Example usage
unique_answers, most_common_answer, most_common_count = count_unique_answers(dataframe, 'King')
print(f'Unique Answers: {unique_answers}')
print(f'Most Common Answer: {most_common_answer}, {most_common_count} times')

````

````python
# Investigate the ways in which questions change over time by filtering by the date. How many questions from the 90s use the word "Computer" compared to questions from the 2000s?

def count_computer_questions(df):
    # Filter for the 90s and 2000s
    df_90s = df[(df['Air Date'] >= '1990-01-01') & (df['Air Date'] < '2000-01-01')]
    df_2000s = df[(df['Air Date'] >= '2000-01-01') & (df['Air Date'] < '2023-01-01')]
    
    # Count the number of questions containing 'Computer'
    count_90s = df_90s['Question'].str.contains('Computer', case=False, na=False).sum()
    count_2000s = df_2000s['Question'].str.contains('Computer', case=False, na=False).sum()
    
    return count_90s, count_2000s

# Example usage:
count_90s, count_2000s = count_computer_questions(dataframe)
print(f'Number of questions from the 90s containing "Computer": {count_90s}')
print(f'Number of questions from the 2000s containing "Computer": {count_2000s}')
````
