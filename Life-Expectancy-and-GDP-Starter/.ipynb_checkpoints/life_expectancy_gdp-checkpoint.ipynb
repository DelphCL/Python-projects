# Goals:

1. Descriptive statistics for life expectancy and GDP by country.

2. Plotting life expectancy and GDP trends over the years for each country.

3. Finding correlations between GDP and life expectancy.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('all_data.csv')

# Display the first few rows of the DataFrame
print(data.head(20))
print(data.describe())
print(data.info())
```

1. Descriptive statistics for life expectancy and GDP by country.

```python
life_expectancy_stats = data.groupby('Country')['Life expectancy at birth (years)'].describe()
gdp_stats = data.groupby('Country')['GDP'].describe()

print('LE', life_expectancy_stats)
print('GDP', gdp_stats)
```

```python

# Assuming the data is already loaded into a DataFrame called df
# ... your DataFrame loading code ...

# List of unique countries
countries = data['Country'].unique()

# For each country, plot the trends
for country in countries:
    # Filter data for the specific country
    subset = data[data['Country'] == country]
    
    # Create a new figure
    fig, ax1 = plt.subplots(figsize=(10, 5))

    # Plot GDP on the primary y-axis (ax1)
    ax1.set_xlabel('Year')
    ax1.set_ylabel('GDP', color='blue')
    ax1.plot(subset['Year'], subset['GDP'], color='blue', label='GDP')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Create a secondary y-axis for life expectancy
    ax2 = ax1.twinx()
    ax2.set_ylabel('Life expectancy at birth (years)', color='green')
    ax2.plot(subset['Year'], subset['Life expectancy at birth (years)'], color='green', label='Life expectancy')
    ax2.tick_params(axis='y', labelcolor='green')

    # Title for the plot
    plt.title(f"{country}'s GDP and Life Expectancy (2000-2015)")

    # Display the plot
    plt.show()
```

3. Finding correlations between GDP and life expectancy.

```python
import pandas as pd
import matplotlib.pyplot as plt

# ... your DataFrame loading code ...

# Calculate correlations for each country
correlations = data.groupby('Country').apply(lambda x: x['GDP'].corr(x['Life expectancy at birth (years)'])).reset_index()
correlations.columns = ['Country', 'Correlation']

# Sort values by correlation for better visualization
correlations = correlations.sort_values(by='Correlation', ascending=False)

# Visualization
plt.figure(figsize=(12, 6))
plt.bar(correlations['Country'], correlations['Correlation'], color='blue', alpha=0.7)
plt.xlabel('Country')
plt.ylabel('Correlation Coefficient')
plt.title('Correlation between GDP and Life Expectancy by Country')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

If the correlation is close to 1, it means that as GDP increases, life expectancy also tends to increase for that country. If the correlation is close to -1, it would mean the opposite. A value close to 0 would indicate little to no linear relationship between the two variables.

```python

import seaborn as sns

# ... your DataFrame loading code ...

# Compute average GDP and life expectancy for each country over the years
data_avg = data.groupby('Country').mean().reset_index()

# Setting up the aesthetic style of the plots
sns.set_style("whitegrid")

# Create scatter plot with adjusted sizes
plt.figure(figsize=(16, 10))

sns.scatterplot(x='GDP', y='Life expectancy at birth (years)', 
                size='GDP', 
                data=data_avg, 
                sizes=(200, 3000),  # Adjusted sizes here
                hue=data_avg['Country'], 
                palette=sns.color_palette("hsv", len(data_avg['Country'])), 
                legend=False)

# Adjusting annotations for each country with offset
for line in range(0, data_avg.shape[0]):
    plt.text(data_avg['GDP'][line]*1.1,  # 10% offset to the right
             data_avg['Life expectancy at birth (years)'][line], 
             data_avg['Country'][line], 
             horizontalalignment='left', 
             size='medium', color='black', backgroundcolor="white")  # Adjusted alignment and offset

# Adjust x-axis to log scale for better visualization
plt.xscale('log')
plt.xlabel('Average GDP (Log Scale)')
plt.title('Average GDP vs. Life Expectancy for Each Country (2000-2015)')
plt.grid(True, which="both", ls="--", c='0.7')

plt.tight_layout()
plt.show()

```
